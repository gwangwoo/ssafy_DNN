```python
import numpy as np
import matplotlib.pylab as plt

def step_func(x):
    return np.array(x>0, dtype=np.int)


x = np.arange(-5.0,5.0,0.1)

y = step_func(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1) # y축의 범위 지정

plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))


x = np.array([-1.0,1.0,2.0])

print(sigmoid(x))

x2 = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x2)

plt.plot(x2,y)
plt.ylim(-0.1,1.1) # y축의 범위지정
plt.show()
```
